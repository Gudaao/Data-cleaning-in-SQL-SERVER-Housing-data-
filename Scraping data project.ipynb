SCRAPING DATA FROM A REAL WEBSITE + PANDAS

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
headers = {
    'User-Agent': 'Scraping data project (contacto: sjpablo6a@gmail.com)'}
page = requests.get(url, headers=headers)
if page.status_code == 200:
    soup = BeautifulSoup(page.text, 'html.parser') # Es mejor usar 'html.parser' o 'lxml'
    print("Título de la página:", soup.title.text)
else:
    print(f"Error al acceder: {page.status_code}")
```

```python
table = soup.find_all('table')[0]
```

```python
world_titles = table.find_all('th')
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
print(world_table_titles)
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    lenght =  len(df)
    df.loc[lenght] = individual_row_data
```
