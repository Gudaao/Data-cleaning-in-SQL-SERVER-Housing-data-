Automatic File Sorter

```python
import os, shutil
```

```python
path = r"C:/Users/57317/Desktop/PROYECTOS PORTAFOLIO/"
```

```python
file_names = os.listdir(path)
print(file_names)
```

```python
folder_names = ["SQL files", "Excel files", "Python files", "Tableu files", "Power BI files"]

for loop in range(0, len(folder_names)):
    if not os.path.exists(path + folder_names[loop]):
        print(path + folder_names[loop])
        os.makedirs(path + folder_names[loop])
        
for file in file_names:
    if ".sql" in file and not os.path.exists(path + "SQL files/" + file):
        shutil.move(path + file, path + "SQL files/" + file)
        print("Se han movido exitosamente archivos de SQL")
    elif ".xlsx" in file and not os.path.exists(path + "Excel files/" + file):
        shutil.move(path + file, path + "Excel files/" + file)
        print("Se han movido exitosamente archivos de Excel")
    elif ".ipynb" in file and not os.path.exists(path + "Python files/" + file):
        shutil.move(path + file, path + "Python files/" + file)
        print("Se han movido exitosamente archivos de Python")
    elif ".twb" in file and not os.path.exists(path + "Tableu files/" + file):
        shutil.move(path + file, path + "Tableu files/" + file)
        print("Se han movido exitosamente archivos de Tableu")
    elif ".pbix" in file and not os.path.exists(path + "Power BI files/" + file):
        shutil.move(path + file, path + "Power BI files/" + file)
        print("Se han movido exitosamente archivos de Power BI")
```
